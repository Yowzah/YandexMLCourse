```python
from sklearn.model_selection import KFold
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import scale
```

```python
with open("wine.data") as file:  
    data = file.read() 
```

```python
data = data.split('\n')
for i in range(len(data)) : data[i] = data[i].split(',')
```

```python
Y = np.array([data[i][0] for i in range(len(data) - 1)])
X = np.array([data[i][1:] for i in range(len(data) - 1)])
```

```python
print(X)
```

```python
kf = KFold(n_splits=5, random_state=42, shuffle=True)
```

```python
max_accur=0
best_k=0
for k in range(1,50):
    neigh = KNeighborsClassifier(n_neighbors=k)
    accur = np.mean(cross_val_score(X=X, y=Y, cv=kf,estimator=neigh))
    if (accur > max_accur):
        max_accur = accur
        best_k = k
```

```python
print(max_accur, best_k)
```

```python
X = scale(X=X)
```

```python
kf = KFold(n_splits=5, random_state=42, shuffle=True)
max_accur=0
best_k=0
for k in range(1,50):
    neigh = KNeighborsClassifier(n_neighbors=k)
    accur = np.mean(cross_val_score(X=X, y=Y, cv=kf,estimator=neigh))
    if (accur > max_accur):
        max_accur = accur
        best_k = k
```

```python
print(max_accur, best_k)
```
